```python
import io

training_file = io.open(r"data/training.txt",encoding='utf-8')

train_sentence = []
train_sentence_string = []
train_label = []

i = 0

for line in training_file:
  line1 = line.split('\t')
  train_label.append(line[0])
  train_sentence.append(line1[1])

print(train_label)
print(len(train_label))

print(train_sentence)
print(len(train_sentence))
```

```python
import pandas as pd

df = pd.DataFrame()

df["Sentences"] = train_sentence
df["Label"] = train_label

print(df)
df.drop_duplicates(inplace=True)
df["Sentences"] = df["Sentences"]
print(df)
```

```python
import numpy as np
import spacy
nlp = spacy.load("en_core_web_md")
tuple_sentences = tuple(df.Sentences)
sentence_vectors = [nlp(sentence).vector for sentence in tuple_sentences]
df['Sentence_vectors'] = sentence_vectors
print(df)
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.cluster import KMeans

X = np.array(df['Sentence_vectors'])
X = np.vstack(X)
y = df['Label']

train_X, test_X, train_y, test_y = train_test_split(X,y,random_state=42,test_size=0.2)
model = SVC(kernel='rbf',random_state=None)
print(train_X.shape)
print(test_X.shape)
model.fit(train_X,train_y)
y_pred = model.predict(test_X)
print('--------------------------------')
print("SVM CLASSIFICATION REPORT:")
print('--------------------------------')
print(classification_report(test_y,y_pred))
print('\n\n')
# model =
# model.fit(train_X,train_y)
# y_pred = model.predict(test_X)
# y_true = np.array(test_y, dtype=int)
# print('--------------------------------')
# print("--- CLASSIFICATION REPORT:")
# print('--------------------------------')
# print(classification_report(y_true,y_pred))
```

```python
from sklearn.decomposition import PCA

number_of_components = 300 # Number of components to retain (your choice)
pca = PCA(n_components=number_of_components)
transformed_data = pca.fit_transform(X)
print(transformed_data.shape)
print(X.shape)

## choose train-test split or hyperparameters accordingly
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
X_train, X_test, y_train, y_test = train_test_split(transformed_data,y,test_size = 0.2, random_state=42)
model = MLPClassifier(hidden_layer_sizes=(number_of_components,300,100,50),activation='tanh',learning_rate='adaptive')
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred))
```
